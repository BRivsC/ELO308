```python
# Ensayo con hilos para mejorar la lectura del puerto serial desde acá
import serial
import threading

puerto_serial = 'COM11'
baud_rate = 115200

def leer_serial():
    with serial.Serial(puerto_serial, baud_rate, timeout=0) as ser:
        while True:
            if ser.in_waiting > 0:
                data = ser.read(ser.in_waiting).decode('ascii')
                print(f"Datos recibidos: {data}")

# Crear un hilo para leer del puerto serial
hilo_serial = threading.Thread(target=leer_serial)
hilo_serial.daemon = True  # El hilo se cierra cuando el programa principal termina
hilo_serial.start()

# Mientras el hilo lee los datos, el programa principal puede hacer otras cosas
try:
    while True:
        pass  # Aquí puedes realizar otras tareas
except KeyboardInterrupt:
    print("Programa terminado.")

# No me gustó el resultado porque no murió bien y la mejora en velocidad tampoco estuvo a la altura del arduino
```

```python
# Ensayo con asyncio
import asyncio
import serial_asyncio

class SerialReader(asyncio.Protocol):
    def connection_made(self, transport):
        self.transport = transport
        print('Connection established')

    def data_received(self, data):
        message = data.decode('utf-8').strip()  # Decodifica el mensaje recibido
        print(f'Received: {message}')           # Imprime el mensaje recibido

    def connection_lost(self, exc):
        print('Connection lost')

async def main():
    loop = asyncio.get_running_loop()

    # Configura la conexión serial en COM11
    transport, protocol = await serial_asyncio.create_serial_connection(
        loop, SerialReader, 'COM11', baudrate=115200
    )

    try:
        await asyncio.sleep(3600)  # Mantiene la conexión durante una hora
    except KeyboardInterrupt:
        pass
    finally:
        transport.close()

# Verifica si ya hay un bucle de eventos corriendo
if __name__ == '__main__':
    try:
        # Crea la tarea principal
        task = asyncio.create_task(main())
        # Deja que el bucle de eventos existente gestione la tarea
    except RuntimeError as e:
        print(f'Error: {e}')

# tampoco es tan rápido como me gustaría
```

```python
# Ensayo usando buffer
import serial

puerto_serial = 'COM11'
baud_rate = 115200
buffer_size = 1024  # Tamaño del buffer, puedes ajustar según tus necesidades

# Configurar el puerto serial
with serial.Serial(puerto_serial, baud_rate, timeout=1) as ser:
    buffer = ""  # Inicializamos un buffer en forma de string para ir acumulando datos
    
    while True:
        if ser.in_waiting > 0:
            # Leer el tamaño del buffer disponible
            data = ser.read(min(buffer_size, ser.in_waiting)).decode('ascii')
            
            # Acumular los datos leídos en el buffer
            buffer += data
            
            # Mientras haya líneas completas (terminadas con un salto de línea)
            while '\n' in buffer:
                # Separar la primera línea completa del buffer
                line, buffer = buffer.split('\n', 1)
                
                # Procesar la línea
                line = line.strip()  # Limpiar espacios en blanco o caracteres extra
                if "CH1:" in line:
                    try:
                        ch1_valor = int(line.split("CH1:")[1].strip())
                        print(f"Registrado CH1: {ch1_valor}")
                    except ValueError:
                        print(f"Error al convertir el dato a entero: {line}")
```

## Ensayo inicial 1 ch + ms desde que empezó a registrar

```python
import serial
import sqlite3
import time
from datetime import datetime

# Configurar el puerto serial
puerto_serial = 'COM11'
baud_rate = 115200

# Conectar o crear la base de datos SQLite
conexion = sqlite3.connect('registro_datos.db')
cursor = conexion.cursor()

# Crear la tabla si no existe
cursor.execute('''CREATE TABLE IF NOT EXISTS datos (
                    id INTEGER PRIMARY KEY AUTOINCREMENT,
                    CH1 INTEGER,
                    timestamp INTEGER,  
                    fecha TEXT,
                    gesto TEXT
                )''')

# Inicializamos la variable de tiempo de referencia (start_time)
start_time = None
gesto = input("Por favor, define el gesto para este registro: ")  # Definir gesto al inicio

# Función para insertar datos en la base de datos
def insertar_dato(ch1_valor):
    global start_time
    
    # Si es el primer dato recibido, inicializamos el tiempo de referencia
    if start_time is None:
        start_time = time.time() * 1000  # Milisegundos

    # Calcular milisegundos transcurridos desde el primer dato
    timestamp = int(time.time() * 1000 - start_time)
    fecha = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

    cursor.execute("INSERT INTO datos (CH1, timestamp, fecha, gesto) VALUES (?, ?, ?, ?)", 
                   (ch1_valor, timestamp, fecha, gesto))
    conexion.commit()

# Abrir el puerto serial y comenzar a leer datos
try:
    with serial.Serial(puerto_serial, baud_rate, timeout=1) as ser:
        print(f"Leyendo desde {puerto_serial} a {baud_rate} baud...\n")
        while True:
            # Leer línea desde el puerto serial
            if ser.in_waiting > 0:
                data = ser.readline().decode('ascii').rstrip()
                
                # Buscar y extraer el valor de CH1
                if "CH1:" in data:
                    try:
                        ch1_valor = int(data.split("CH1:")[1].strip())
                        insertar_dato(ch1_valor)
                        print(f"Registrado CH1: {ch1_valor}")
                    except ValueError:
                        print(f"Error al convertir el dato a entero: {data}")
except serial.SerialException as e:
    print(f"Error al acceder al puerto serial: {e}")
except KeyboardInterrupt:
    print("\nLectura interrumpida. Saliendo...")

# Cerrar la conexión a la base de datos
finally:
    conexion.close()
```

# Procesamiento de datos
En esta parte se le hacen consultas a la base de datos para extraerlos y graficarlos. 

```python
import sqlite3
import matplotlib.pyplot as plt
import datetime
import os

# Conectar a la base de datos SQLite
db_path = 'registro_datos_1ch.db'
conexion = sqlite3.connect(db_path)
cursor = conexion.cursor()

# Mostrar la ubicación de la base de datos en consola
print(f"Usando base de datos en: {os.path.abspath(db_path)}")

# Función para obtener datos de un gesto específico
def obtener_datos_gesto(gesto_id):
    cursor.execute("SELECT timestamp, CH1_fil, CH1_env FROM datos WHERE gesto_id = ?", (gesto_id,))
    datos = cursor.fetchall()
    return datos

# Pedir el gesto_id del gesto a graficar
gesto_id = int(input("Por favor, introduce el gesto_id a graficar: "))

# Obtener los datos del gesto
datos = obtener_datos_gesto(gesto_id)

# Separar los datos en listas para graficar
timestamps = [dato[0] for dato in datos]
CH1_fil_values = [dato[1] for dato in datos]
CH1_env_values = [dato[2] for dato in datos]

# Graficar los datos
plt.figure(figsize=(12, 6))

plt.plot(timestamps, CH1_fil_values, label='CH1 Filtrado', color='blue')
plt.plot(timestamps, CH1_env_values, label='CH1 Envolvente', color='red', linestyle='--')

plt.xlabel('Timestamp (ms)')
plt.ylabel('Valor')
plt.title('Datos de EMG para el gesto con gesto_id = {}'.format(gesto_id))
plt.legend()

# Mostrar la ubicación de la base de datos en consola
print(f"Usando base de datos en: {os.path.abspath(db_path)}")

# Mostrar la gráfica
plt.show()

# Cerrar la conexión a la base de datos
conexion.close()
```
